## 2. Set up the knowledge base

```python
import os
import langchain.chains

from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.text_splitter import CharacterTextSplitter
from langchain import OpenAI, VectorDBQA
from langchain.document_loaders import DirectoryLoader
from langchain.chains import RetrievalQA
 
# OpenAI API key
os.environ["OPENAI_API_KEY"] = 'YOUR-OPENAI-KEY'
 
# load all the txt files in the directory
# in case of other formats, change the glob parameter
loader = DirectoryLoader(r'D:\mydocument\juxue\psychological_data\book\book_cn', glob='**/*.txt')

# change data into document, each file will be a document
documents = loader.load()
 
# initialization the text splliter
text_splitter = CharacterTextSplitter(chunk_size=400, chunk_overlap=0)

# split the document into chunks
split_docs = text_splitter.split_documents(documents)
 
# initialization the openai embeddings object
embeddings = OpenAIEmbeddings()

# put the document into the Chroma vector database for further matching query
# the vector_store_path is the path to store the vector database
# here we set the persist directory as vector_store_path
vector_store_path = r"D:\mydocument\juxue\vector_knowledge_base"
docsearch = Chroma.from_documents(documents=split_docs,
                                  embedding=embeddings,
                                  persist_directory=vector_store_path)
 
# docsearch = Chroma(persist_directory=vector_store_path,embedding_function=embeddings)

# create a retrieval question answering system
qa = VectorDBQA.from_chain_type(llm=OpenAI(), chain_type="stuff", vectorstore=docsearch, return_source_documents=True)
```

```python
# 进行问答
result = qa({"query": "今晚吃什么？"})
print(result)
```
